```python
import requests
from xml.etree import ElementTree as ET
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import os
import easyocr
import numpy as np
import datetime

# %matplotlib inline


class Coordinate:
    def __init__(self, lat: float, lon: float):
        # ensure that the latitude is between -90 and 90
        if lat < -90 or lat > 90:
            raise ValueError("Latitude must be between -90 and 90")
        # ensure that the longitude is between -180 and 180
        if lon < -180 or lon > 180:
            raise ValueError("Longitude must be between -180 and 180")

        self.lat = lat
        self.lon = lon

    def __str__(self):
        return f"Lat: {self.lat}, Lon: {self.lon}"


# Meters per second to knots
def MpsToKts(mps: float) -> float:
    return mps * 1.94384


class BuoyObservation:
    def __init__(
        self,
        station_id: str,
        timestamp: datetime.datetime,
        coordinates: Coordinate,
        wind_speed_kts: float = None,
        wind_direction_deg: float = None,
        wind_gust_kts: float = None,
        wave_height_m: float = None,
        wave_direction_deg: float = None,
    ):
        self.station_id: int = station_id
        self.timestamp: datetime.datetime = timestamp
        self.coordinates: Coordinate = coordinates
        self.wind_speed_kts: float = wind_speed_kts
        self.wind_direction_deg: float = wind_direction_deg
        self.wind_gust_kts: float = wind_gust_kts
        self.wave_height_m: float = wave_height_m
        self.wave_direction_deg: float = wave_direction_deg

    def __str__(self):
        return f"Station: {self.station_id}, Timestamp: {self.timestamp}, Coordinates: {self.coordinates}, Wind Speed: {self.wind_speed_kts}, Wind Direction: {self.wind_direction_deg}, Wind Gust: {self.wind_gust_kts}, Wave Height: {self.wave_height_m}, Wave Direction: {self.wave_direction_deg}"


def extract_date_string(filename):
    # Find the position of the first underscore
    start_index = filename.find("_") + 1
    if start_index == -1:
        return None
    # The end index is where ".jpg" starts
    end_index = filename.find(".jpg")
    if end_index == -1:
        return None
    # Extract the substring from start_index to end_index
    return filename[start_index:end_index]


def date_string_to_date(date_string):
    # Extract the year, month, and day from the date string. All times are in UTC.
    year = int(date_string[:4])
    month = int(date_string[5:7])
    day = int(date_string[8:10])
    hour = int(date_string[11:13])
    minute = int(date_string[13:15])
    return datetime.datetime(year, month, day, hour, minute, tzinfo=datetime.timezone.utc)


def extract_csv(url):
    response = requests.get(url)
    # verify that the request was successful
    response.raise_for_status()

    data = response.text
    lines = data.split("\n")

    # Verify that the file has at least 3 lines (header, units, and data)
    if len(lines) < 3:
        raise ValueError("File does not have enough lines")

    # Extract the header and the first row
    header = lines[0].split()
    # Skip the second line which is the units
    rows = lines[2:]
    # Create a diciotnary where the keys are the station numbers which is the first column of each row.
    # The values are a dictionary of the header and row values
    result = {}
    for row in rows:
        values = row.split()
        if len(values) == 0:
            continue
        station_number = values[0]
        result[station_number] = {}
        for i in range(1, len(header)):
            result[station_number][header[i]] = values[i]

    return result


def create_buoy_observations(data):
    observations = {}

    for station_number, station_data in data.items():
        date = datetime.datetime(
            int(station_data["YYYY"]),
            int(station_data["MM"]),
            int(station_data["DD"]),
            int(station_data["hh"]),
            int(station_data["mm"]),
            tzinfo=datetime.timezone.utc,
        )
        coordinates = Coordinate(float(station_data["LAT"]), float(station_data["LON"]))
        # verify if station_data has an entry for WSPD and it is not MM (missing data)
        wind_speed_kts = (
            MpsToKts(float(station_data["WSPD"])) if "WSPD" in station_data and station_data["WSPD"] != "MM" else None
        )
        wind_direction_deg = (
            float(station_data["WDIR"]) if "WDIR" in station_data and station_data["WDIR"] != "MM" else None
        )
        wind_gust_kts = (
            MpsToKts(float(station_data["GST"])) if "GST" in station_data and station_data["GST"] != "MM" else None
        )
        wave_height_m = float(station_data["WVHT"]) if "WVHT" in station_data and station_data["WVHT"] != "MM" else None
        wave_direction_deg = (
            float(station_data["MWD"]) if "MWD" in station_data and station_data["MWD"] != "MM" else None
        )
        observation = BuoyObservation(
            station_number,
            date,
            coordinates,
            wind_speed_kts,
            wind_direction_deg,
            wind_gust_kts,
            wave_height_m,
            wave_direction_deg,
        )
        observations[station_number] = observation
    return observations


NUM_SMALL_IMAGES_PER_ROW = 6
DEGREES_PER_IMAGE = 360 / NUM_SMALL_IMAGES_PER_ROW
BUOYCAMS_KML_SOURCE = "https://www.ndbc.noaa.gov/kml/buoycams_as_kml.php"
LATEST_OBSERVATION_URL = "https://www.ndbc.noaa.gov/data/latest_obs/latest_obs.txt"

# fetch data from LATEST_OBSERVATION_URL

latest_observation_data = extract_csv(LATEST_OBSERVATION_URL)
observations = create_buoy_observations(latest_observation_data)

# # print the observations
# for observation in observations:
#     print(observation)

# Iterate through the dictionary and print the station number and the station data. Print each station's data on a single line

# fetch data from BUOYCAMS_KML_SOURCE
response = requests.get(BUOYCAMS_KML_SOURCE)
data = response.text
root = ET.fromstring(data)

root_attributes = root.attrib

namespace_uri = root.tag[root.tag.find("}") + 1 :]  # This extracts everything after the '}' which is the tag name
namespace = root.tag[: root.tag.find("}")]  # This extracts the namespace URL including the brackets {}
namespace = namespace.strip("{").strip("}")  # Clean up to remove the brackets

ns = {namespace_uri: namespace}

print(f"Namespace: {ns}")

# Extract document name
doc_name = root.find(".//kml:Document/kml:name", ns)
print(f'Document Name: {doc_name.text if doc_name is not None else "Unnamed Document"}')

ocr_reader = easyocr.Reader(["en"])

# Iterate over all Placemarks
placemarks = root.findall(".//kml:Placemark", ns)


for placemark in placemarks:
    name = placemark.find("kml:name", ns)
    description = placemark.find("kml:description", ns)
    coordinates = placemark.find(".//kml:coordinates", ns)

    assert name is not None

    station_name = name.text

    print(f"Station Name: {station_name}")
    print(f'Description: {description.text if description is not None else "No Description"}')
    print(f'Coordinates: {coordinates.text.strip() if coordinates is not None else "No Coordinates"}')

    # extras = placemark.find(".//kml:LookAt", ns)
    # if extras is not None:
    #     for extra in extras:
    #         # for extras.tag print everything after the final '}'
    #         tag = extra.tag[extra.tag.find('}')+1:]
    #         print(f'{tag}: {extra.text}')

    # Extract the image URL fro the description text if it exists
    img_url = None
    if description is not None:
        if '<img src="' in description.text:
            img_url = description.text.split('<img src="')[1].split('"')[0]
            print(f"Image URL: {img_url}")

    if img_url is None:
        continue
    # Download the image
    img_response = requests.get(img_url)

    # Display the image with matplotlib

    img = Image.open(BytesIO(img_response.content))
    plt.title(f"Station: {station_name}, Date: {img_date}")
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    full_img_size = img.size
    print(f"Image Size: {full_img_size}")
    full_img_width, full_img_height = full_img_size
    img_format = img.format

    # split the image into 6 480x270 images
    small_img_width, small_img_height = (480, 270)

    # Extract the image angle from the bottom left corner of the image data
    bottom_left_img = img.crop((150, full_img_height - 30, 250, full_img_height))

    results = ocr_reader.readtext(np.array(bottom_left_img))

    first_image_angle = None

    if len(results) > 0:
        # Remove the last character (degree symbol) from the result
        first_result = results[0][1][:-1]
        # verify that the result is a number
        if first_result.isnumeric():
            first_image_angle = int(first_result)

    assert first_image_angle is not None
    if first_image_angle is None:
        first_image_angle = 0

    # Get the image name from the URL
    img_filename = img_url.split("/")[-1]
    img_name = img_filename.split(".")[0]

    img_parts = []
    for i in range(6):
        left = i * small_img_width
        angle = int((first_image_angle + i * DEGREES_PER_IMAGE) % 360)
        sub_img = img.crop((left, 0, left + small_img_width, small_img_height))
        # plt.title(f"Station: {station_name}, Date: {img_date}, Angle: {angle}")
        # plt.imshow(sub_img)
        # plt.axis("off")
        # plt.show()
        img_parts.append({"image": sub_img, "angle": angle})

    # Get the date from the image filename
    img_date_str = extract_date_string(img_filename)
    img_date = date_string_to_date(img_date_str)
    print(f"Image Date: {img_date}")

    station_name = name.text

    img_age = datetime.datetime.now(datetime.timezone.utc) - img_date
    print(f"Image Age: {img_age}")

    assert observations[station_name] is not None

    print(observations[station_name])

    observation = observations[station_name]

    # Compare the img_date with now
    observation_age = datetime.datetime.now(datetime.timezone.utc) - observation.timestamp
    print(f"Observation Age: {observation_age}")

    observation_image_age_diff = observation.timestamp - img_date
    print(f"Observation Image Age Difference: {observation_image_age_diff}")

    # make the folder if it doesn't exist
    dir = f"images/{img_date_str}/{station_name}"
    if not os.path.exists(dir):
        os.makedirs(dir)

    # Save the full image to disk in an images folder
    full_img_path = f"{dir}/{img_filename}"
    img.save(full_img_path)
    print(f"Full image saved as {full_img_path}")

    # Same the image parts to disk
    for img_part in img_parts:
        img_part_filename = f'{img_name}_{img_part["angle"]}.jpg'
        img_part_path = f"{dir}/{img_part_filename}"
        img_part["image"].save(img_part_path)
        # print(f'Image saved as {img_part_path}')
```
