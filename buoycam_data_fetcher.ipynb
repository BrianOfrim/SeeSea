```python
import requests
from xml.etree import ElementTree as ET
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import re
import os
import easyocr
import numpy as np

# %matplotlib inline


def extract_date_string(filename):
    # Find the position of the first underscore
    start_index = filename.find("_") + 1
    if start_index == -1:
        return None
    # The end index is where ".jpg" starts
    end_index = filename.find(".jpg")
    if end_index == -1:
        return None
    # Extract the substring from start_index to end_index
    return filename[start_index:end_index]


NUM_SMALL_IMAGES_PER_ROW = 6
DEGREES_PER_IMAGE = 360 / NUM_SMALL_IMAGES_PER_ROW
BUOYCAMS_KML_SOURCE = "https://www.ndbc.noaa.gov/kml/buoycams_as_kml.php"
# fetch data from BUOYCAMS_KML_SOURCE
response = requests.get(BUOYCAMS_KML_SOURCE)
data = response.text
root = ET.fromstring(data)

root_attributes = root.attrib

namespace_uri = root.tag[root.tag.find("}") + 1 :]  # This extracts everything after the '}' which is the tag name
namespace = root.tag[: root.tag.find("}")]  # This extracts the namespace URL including the brackets {}
namespace = namespace.strip("{").strip("}")  # Clean up to remove the brackets

ns = {namespace_uri: namespace}

print(f"Namespace: {ns}")

# Extract document name
doc_name = root.find(".//kml:Document/kml:name", ns)
print(f'Document Name: {doc_name.text if doc_name is not None else "Unnamed Document"}')

ocr_reader = easyocr.Reader(["en"])

# Iterate over all Placemarks
placemarks = root.findall(".//kml:Placemark", ns)
for placemark in placemarks:
    name = placemark.find("kml:name", ns)
    description = placemark.find("kml:description", ns)
    coordinates = placemark.find(".//kml:coordinates", ns)

    print(f'Station Name: {name.text if name is not None else "Unnamed"}')
    print(f'Description: {description.text if description is not None else "No Description"}')
    print(f'Coordinates: {coordinates.text.strip() if coordinates is not None else "No Coordinates"}')

    # extras = placemark.find(".//kml:LookAt", ns)
    # if extras is not None:
    #     for extra in extras:
    #         # for extras.tag print everything after the final '}'
    #         tag = extra.tag[extra.tag.find('}')+1:]
    #         print(f'{tag}: {extra.text}')

    # Extract the image URL fro the description text if it exists
    img_url = None
    if description is not None:
        if '<img src="' in description.text:
            img_url = description.text.split('<img src="')[1].split('"')[0]
            print(f"Image URL: {img_url}")

    if img_url is None:
        continue
    # Download the image
    img_response = requests.get(img_url)

    # Display the image with matplotlib

    img = Image.open(BytesIO(img_response.content))
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    full_img_size = img.size
    print(f"Image Size: {full_img_size}")
    full_img_width, full_img_height = full_img_size
    img_format = img.format

    # split the image into 6 480x270 images
    small_img_width, small_img_height = (480, 270)

    # Extract the image angle from the bottom left corner of the image data

    bottom_left_img = img.crop((150, full_img_height - 30, 250, full_img_height))

    results = ocr_reader.readtext(np.array(bottom_left_img))

    first_image_angle = None

    if len(results) > 0:
        # Remove the last character (degree symbol) from the result
        first_result = results[0][1][:-1]
        # verify that the result is a number
        if first_result.isnumeric():
            first_image_angle = int(first_result)

    assert first_image_angle is not None
    if first_image_angle is None:
        first_image_angle = 0

    # Get the image name from the URL
    img_filename = img_url.split("/")[-1]
    img_name = img_filename.split(".")[0]

    img_parts = []
    for i in range(6):
        left = i * small_img_width
        angle = int((first_image_angle + i * DEGREES_PER_IMAGE) % 360)
        sub_img = img.crop((left, 0, left + small_img_width, small_img_height))
        plt.title(f"Station: {station_name}, Date: {img_date}, Angle: {angle}")
        plt.imshow(sub_img)
        plt.axis("off")
        # plt.show()
        img_parts.append({"image": sub_img, "angle": angle})

    # Get the date from the image filename
    img_date = extract_date_string(img_filename)
    print(f"Image Date: {img_date}")

    station_name = name.text

    # make the folder if it doesn't exist
    dir = f"images/{img_date}/{station_name}"
    if not os.path.exists(dir):
        os.makedirs(dir)

    # Save the full image to disk in an images folder
    full_img_path = f"{dir}/{img_filename}"
    img.save(full_img_path)
    print(f"Full image saved as {full_img_path}")

    # Same the image parts to disk
    for img_part in img_parts:
        img_part_filename = f'{img_name}_{img_part["angle"]}.jpg'
        img_part_path = f"{dir}/{img_part_filename}"
        img_part["image"].save(img_part_path)
        # print(f'Image saved as {img_part_path}')
```
